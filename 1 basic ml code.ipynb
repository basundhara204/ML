```python
# importing all required libraries and modules
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
```

```python
# loadind dataset into variable and display in dataframe
iris=load_iris()
df=pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species']=iris.target
df['species']=df['species'].map(dict(enumerate(iris.target_names)))
df.head()  # for first five rows  and 'df' for all rows
```

```python
# loadind dataset into variable and runnning the model
iris=load_iris()
x=iris.data   #mearurement 
y=iris.target #target
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2 , random_state=42)
model = RandomForestClassifier()
model.fit(x_train, y_train)
y_pred= model.predict(x_test)
print("accuracy", accuracy_score(y_test, y_pred))
```
